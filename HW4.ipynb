## 4.1. Работа с датасетом "house price"
1.1 Скачать данные по ссылке kc-house-data и laptop_price


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc-house-data (1).csv', encoding='cp1251')
df.head()
```

1.2 Изучите стоимости недвижимости

```python
plt.figure()
plt.hist(df['price'])
plt.xlabel('Стоимость недвижимости')
plt.ylabel('Кол-во домов')
plt.title('Распределение стоимости недвижимости')
```

На основании графика "Распределение стоимости недвижимости" можно сделать вывод,  что большее распределение получилили дома с относительно низкой стоимостью, а домов по средней и высокой стоимости значительно меньше.

1.3 Изучите распределение квадратуры жилой площади

```python
plt.figure()
plt.hist(df['sqft_living'], bins=500)
plt.xlabel('Квадратные метры')
plt.ylabel('Кол-во домов')
plt.title('Распределение квадратуры жилой площади')
```

На основании графика распределения площади недвижимости, можно сделать вывод, что больше всего домов площадью в пределах 1800 - 2500 кв. м. Недвижимости площадью от 6000 кв. м минимальное количество.

1.4 Изучите распределение года постройки
- Постройте график
- Назовите график
- Сделайте именование оси x и оси y

```python
plt.figure()
plt.hist(df['yr_built'], bins=50)
plt.xlabel('Год постройки')
plt.ylabel('Кол-во домов')
plt.title('Распределение года постройки')
```

Гистограмма распределения недвижимости относительно года постройки даёт понять, что домов "свежих" лет постройки больше , чем старых.Больше всего построили домов в период с 2005-2010г. В 1930 -1937 меньше всего строили новых объектов, вероятно из-за катаклизмов и военных действий.Далее периодически постройки новых домой возобновляются раз в несколько лет.И также после массовой постройки новых объектов(пики подъемов) несколько лет после происходит пауза в строительстве.

4.3 Работа с датасетом "house price"
4.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
dCorr = df.corr(numeric_only=True).sort_values('price')
dCorr = np.round(dCorr, 2)
dCorr = np.abs(dCorr)
dCorr
```

Чтобы определить какие параметры больше всего влияют на стоимость недвижимости, в таблице корелляции возьмем 5 наибольших значений - sqft_living (0,70), grade (0,67), sqft_above(0,61), bathrooms (0,53), view (0,40). Это означает, что размер жилой площади, качество консрукции и дизайна, квадратные метры, которые находятся выше земли, количество ванных комнат и хороший вид очень сильно влияют на стоимость недвижимости. Рассмотри это на следущих графиках и диаграммах:

```python
plt.figure()
plt.hist(df['price'])
plt.xlabel('price')
plt.ylabel('Кол-во домов')
plt.title('Распределение цены')
```

```python
sns.histplot(data = df, x = 'price', hue ='view', bins= 30)
```

На диаграмме зависимости цены от качества вида можно сделать вывод, что большинство домов имеют вид на оценку от 0 до 1, а домов  с отличным видом мало, и им соответствует высокая цена.

```python
living = df['sqft_living'].value_counts()
living
```

```python
plt.figure()
sns.scatterplot(x=df['sqft_living'], y=df['price'])
plt.title('График зависимости стоимости от жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость недвижимости');
```

По данному графику видно, что цена растет с размером жилой площади.Самая большая плотность домов площадью примерно до 5500 м кв.

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

Анализ взаимного распределения оценки уровня конструкци и дизайна с ценой показывает, что более высокая оценка подразумевает рост цены на такую недвижимость.

```python
plt.figure()
plt.hist(df['sqft_above'])
plt.xlabel('Площадь')
plt.ylabel('Кол-во домов')
plt.title('Распределение площади')
```

```python
plt.figure()
plt.plot(df['sqft_living'], df['price'])
plt.title('Динамика площади, которая находитмя выше земли по отношению к цене')
plt.xlabel('Площадь')
plt.ylabel('Цена недвижимости');
```

Данный график показывает, дома одинаковые по площади можно купить по разной стоимости(это зависит от состояния дома, вида и т.д). И чем больше растет площадь дома над уровнем земли тем выше и разброс цен на жилье.Чем меньше дом и ментше площадь, тем меньше выбора по стоимости при его поиске.
