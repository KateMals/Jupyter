- Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)

- Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)

- Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)

- Удалите признаки Memory и ScreenResolution

```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='cp1251')
```

```python
df.head(5)
```

```python
def vstavka(stroka):
    return stroka.split()[0]
```

```python
df['Cpu_Company'] = df['Cpu'].apply(vstavka)
```

```python
df.head()
```

```python
df['Cpu_Company'].value_counts()
```

```python
def sumMemory(stro4ka):
    return int(float(stro4ka.split()[0][:-2]))
```

```python
df['Memory_Amount'] = df['Memory'].apply(sumMemory)
```

```python
df['Memory_Amount'].value_counts()
```

```python
sumMemory('256GB SSD +  1.0TB Hybrid')
```

```python
df['Memory_Amount'] = df['Cpu'].apply(vstavka)
```

```python
df.head()
```

```python
def Tip(memorytype):
    return memorytype.split()[1]
```

```python
df['Memory_Type'] = df['Memory'].apply(Tip)
```

```python
df['Memory_Type'].value_counts()
```

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```

```python
df['SSD'] = 0
df.head()
```

```python
df.loc[df['Memory_Type'] == 'SSD','SSD'] = 1
df.head()
```

```python
df['SSD'].value_counts()
```
